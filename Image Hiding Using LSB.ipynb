```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=np.array(img)
```

```python
w,h,_=img.shape
```

```python
message=input()
message += "[END]"
message = message.encode("ascii")
message_bits =''.join([format(i,'08b') for i in message])
```

```python
img=img.flatten()
for idx,bit in enumerate(message_bits):
    val=img[idx]
    val=bin(val)
    val=val[:-1]+bit
    img[idx]=int(val,2)
img = img.reshape((w,h,_))
```

```python
plt.imshow(img)
```

## Decode the image

```python
msg=""
img=img.flatten()
idx=0
while msg[-5:] != '[END]':
    bits=[bin(i)[-1] for i in img[idx:idx+8]]
    bits=''.join(bits)
    msg += chr(int(bits,2))
    idx += 8
    if idx > img.shape[0]-8:
        print("There is no hidden message.")
        break
```
